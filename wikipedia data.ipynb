```python
import numpy as np
import pandas as pd 
```

```python
from pandas.io.html import read_html
```

```python
page = 'https://en.wikipedia.org/wiki/List_of_countries_by_hospital_beds'

wikitables = read_html(page,attrs={"class":"wikitable"})

print ("Extracted {num} wikitables".format(num=len(wikitables)))
```

```python
wikitables[0].head()
```

```python
wikitables[0].drop(0,axis=1,inplace=True)
```

```python
wikitables[0].drop([1,2],axis=0,inplace=True)
```

```python
wikitables[0].drop([],axis=0,inplace=True)
```

```python
wikitables[0].head()
```

```python
#df.rename(columns={'A': 'a'}, index={'ONE': 'one'})
wikidata=wikitables[0].rename(columns={1:'Country',
                            2:'Continent',
                            3:'2013',
                            4:'2014',
                            5:'2015',
                            6:'2016',
                            7:'2017',
                            8:'Actual',
                            9:'Proportional (%)',
                            10:'Occupancy (%)',
                            11:'ICU-CCB beds',
                            12:'Ventilators'})
```

```python
wikidata.to_csv('wikistatic.csv')
```
