```python
#insert package
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plot 
import matplotlib.style as style
%matplotlib inline
```

```python
#plot style 
sns.set(rc={'figure.figsize':(10,7)})
style.use('ggplot')
sns.set_context('paper')
```

```python
#read data file
covdata=pd.read_excel('BNO Data-March 21, 2020.xlsx')
```

```python
#sort data by cases number 
covdata.sort_values('Cases',axis=0,ascending=False)
```

```python
#select first 10th countries
f10=covdata[:10]
```

```python
f10.drop([3],inplace=True,axis=0)
```

```python
#add japan row to selected data frame
f10=f10.append(covdata[covdata['WORLD']=='Japan'],ignore_index=True)
```

```python
f10=f10.sort_values('% of deaths',axis=0,ascending=False)
f10.head(12)
```

```python
chart1=sns.barplot(x='% of deaths',y='WORLD',data=f10,palette="plasma")
chart1.set(xlabel="% of deaths untill March 21, 2020", ylabel = "Countries")
plot.xlim(0.0,0.1)
chart1.tick_params(axis='both', which='both', length=0)
total =len(f10['% of deaths'])

for p in chart1.patches:
    percentage = '{:.2f}%'.format(100* (p.get_width()))
    x = p.get_x() + p.get_width() + 0.001
    y = p.get_y() + p.get_height()/1.5
    chart1.annotate(percentage, (x, y))
    
```

```python
chart2=sns.barplot(x='Cases',y='WORLD',data=f10.sort_values('Cases',ascending=False),palette="plasma")
plot.title('Chart 2: Number of Cases As of 21 March')
chart2.set(xlabel="number of cases untill March 21, 2020", ylabel = "Countries")
plot.xlim(100,90000)
chart2.tick_params(axis='both', which='both', length=0)
total =len(f10['Cases'])
plot.xlabel('Number of Cases As of 21 March', fontsize = 15, weight = 'bold')

#display value on bar 
for p in chart2.patches:
    percentage = '{}'.format((p.get_width()))
    x = p.get_x() + p.get_width() + 500
    y = p.get_y() + p.get_height()/1.5
    chart2.annotate(percentage, (x, y))
```

```python
def count_actual(x,y,z):
    country=x
    num= float(y)
    Actual= float(z)
    
    if country != 'South Korea':
        return num-(Actual*2)
    elif country == 'South Korea':
        return num+(Actual*2)
    else:
        return 0
    
```
